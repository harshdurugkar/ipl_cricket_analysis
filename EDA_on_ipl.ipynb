```python
import pandas as pd
from matplotlib import pyplot as plt
```

#                              EDA by Harsh Durugkar on IPL 2017 Season

```python
ipldata = pd.read_csv('matches.csv') 
ipldata.head()
```

```python
ipldata['player_of_match'].value_counts()[:5]
```

```python
plt.figure(figsize=(10,5))
plt.bar(list(ipldata['player_of_match'].value_counts()[:5].keys()),ipldata['player_of_match'].value_counts()[:5])
plt.title("Bar Graph Shows Top 5 Highest Man of the Match Players in 2017 Season" , size=25, c='g')
plt.xlabel("Player Name" , size = 20 )
plt.ylabel("No. of times " , size = 20)
plt.show()
```

```python
plt.figure(figsize=(23,5))
plt.bar(list(ipldata['toss_winner'].value_counts().keys()),ipldata['toss_winner'].value_counts(),color=["SkyBlue","Red","#7E2F8E","Blue","DarkOrange","#c720fa","#fc7e00","#fc3700"])
plt.title("Bar Graph Shows Top Team to win most toss in 2017 Season" , size=25, c='g')
plt.xlabel("Team Name" , size = 20 )
plt.ylabel("No. of times " , size = 20)
plt.show()
```

```python
ipldata['After_decision_match_result'].value_counts()[:5]
```

```python
plt.figure(figsize=(2,3))
plt.bar(list(ipldata['After_decision_match_result'].value_counts().keys()),ipldata['After_decision_match_result'].value_counts(),color=["#11de07","#fa2020"])
plt.title("Bar Graph Shows is won the toss benifted or not in 2017 Season" , size=25, c='g')
plt.xlabel("After Toss" , size = 20 )
plt.ylabel("No. of times " , size = 20)
plt.show()
```

```python
plt.figure(figsize=(2,3))
plt.bar(list(ipldata['toss_decision'].value_counts().keys()),ipldata['toss_decision'].value_counts())
plt.title("Bar Graph Shows the Most Selected Toss decision in 2017 Season" , size=25, c='g')
plt.xlabel("After Toss" , size = 20 )
plt.ylabel("No. of times " , size = 20)
plt.show()
```

```python
plt.figure(figsize=(23,5))
plt.bar(list(ipldata['winner'].value_counts().keys()),ipldata['winner'].value_counts(),color=["SkyBlue","#c720fa","#7E2F8E","#fc7e00","#fc3700","Blue","DarkOrange","Red"])
plt.title("Bar Graph Shows the Most Match Winner in 2017 Season" , size=25, c='g')
plt.xlabel("Team Name" , size = 20 )
plt.ylabel("No. of times " , size = 20)
plt.show()
```

```python
plt.figure(figsize=(23,5))
plt.bar(list(ipldata['city'].value_counts().keys()),ipldata['city'].value_counts(),color=["SkyBlue","#c720fa","#7E2F8E","#fc7e00","#fc3700","Blue","DarkOrange","Red"])
plt.title("Bar Graph Shows which city most organised the match " , size=25, c='g')
plt.xlabel("City Name" , size = 20 )
plt.ylabel("No. of times " , size = 20)
plt.show()
```
